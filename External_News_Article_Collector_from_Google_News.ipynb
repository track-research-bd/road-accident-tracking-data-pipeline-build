```python
from serpapi import GoogleSearch
import pandas as pd

params = {
    "api_key": "",       #put api
    "q": "bangladesh road accident",
    "tbm": "nws",
    "hl":"en",
    "lr":"en",
}

search = GoogleSearch(params)
pages = search.pagination()
n=0
upperframe=[]
frame=[]
for page in pages:
    for result in page["news_results"]:
      if "ago" not in result['date']:
        frame.append((result['date'],result['link'],result['title']))
    #n=n+1
    #if n==2:
      #break
upperframe.extend(frame)
collection=pd.DataFrame(upperframe, columns=['date','url','title'])
collection
```

```python
from google.colab import files
collection.to_csv('newsarticles.csv', encoding = 'utf-8-sig')
files.download('newsarticles.csv')
```
