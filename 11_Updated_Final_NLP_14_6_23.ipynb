```python
def extract_article_from_url(url):
  a = Article(url)
  a.download()
  a.parse()
  return a.text, a.publish_date
```

```python
def sentence_giver(doc, start, end):
  s = start
  while str(doc[s]) != '.':
    s = s - 1
    if s < 0:
      break
  e = end
  while str(doc[e]) != '.':

    e = e + 1

    if e > len(doc) - 1:
      break

  return str(doc[s+1:e])
```

```python
def location_finder(doc):
  locations = []
  sentences = []
  matcher = Matcher(nlp.vocab)
  pattern = [{"ENT_TYPE": "GPE"}]
  matcher.add("Location", [pattern])
  matches = matcher(doc)
  for match_id, start, end in matches:
    sentence = sentence_giver(doc, start, end)
    value = l_detail_filter(sentence)
    if value == 1:
      sentences.append(sentence)
      locations.append(str(doc[start:end]))
  return locations, sentences
```

```python
def l_detail_filter(data):
  data = nlp(data)
  matcher = Matcher(nlp.vocab)
  pattern0 = [{"LOWER": "in"}]
  pattern1 = [{"LEMMA": "accident"}, {"LOWER": "in"}]
  pattern2 = [{"LEMMA": "die"}, {"LOWER": "in"} ]
  pattern3 = [{"LEMMA": "collide"}, {"LOWER": "in"} ]
  pattern4 = [{"LEMMA": "crash"}, {"LOWER": "in"} ]
  pattern5 = [{"LEMMA": "kill"}, {"LOWER": "in"} ]

  matcher.add("L_Filter", [pattern1])
  matches = matcher(data)
  if len(matches) != 0:
    return 1
  else:
    matcher.add("L_Filter", [pattern2])
    matches = matcher(data)
    if len(matches) != 0:
      return 1
    else:
      matcher.add("L_Filter", [pattern3])
      matches = matcher(data)
      if len(matches) != 0:
        return 1
      else:
        matcher.add("L_Filter", [pattern4])
        matches = matcher(data)
        if len(matches) != 0:
          return 1
        else:
          matcher.add("L_Filter", [pattern5])
          matches = matcher(data)
          if len(matches) != 0:
            return 1
          else:
            matcher.add("L_Filter", [pattern0])
            matches = matcher(data)
            if len(matches) != 0:
              return 1
            else:
              return 0

```

```python
def number_finder(doc):
  numbers = []
  sentences = []
  matcher = Matcher(nlp.vocab)
  pattern = [{"LIKE_NUM": True}]
  matcher.add("Number", [pattern])
  matches = matcher(doc)
  for match_id, start, end in matches:
    sentence = sentence_giver(doc, start, end)
    value = n_detail_filter(sentence)
    if value == 1:
      sentences.append(sentence)
      numbers.append(str(doc[start:end]))
  return numbers, sentences
```

```python
def n_detail_filter(data):
  data = nlp(data)
  matcher = Matcher(nlp.vocab)
  pattern1 = [{"LEMMA": "kill"}]
  pattern2 = [{"LEMMA": "dead"}]
  pattern3 = [{"LEMMA": "die"}]


  matcher.add("N_Filter", [pattern1])
  matches = matcher(data)
  if len(matches) != 0:
    return 1
  else:
    matcher.add("N_Filter", [pattern2])
    matches = matcher(data)
    if len(matches) != 0:
      return 1
    else:
      matcher.add("N_Filter", [pattern3])
      matches = matcher(data)
      if len(matches) != 0:
        return 1
      else:
        return 0

```
